```python
pm10=pm10[pm10["Sample Duration"]=="24-HR BLK AVG"]\
    .rename(columns={"Arithmetic Mean":"PM10"}).drop(columns=["Sample Duration"])
pm25_frm_fem=pm25_frm_fem[pm25_frm_fem["Sample Duration"]=="24-HR BLK AVG"]\
    .rename(columns={"Arithmetic Mean":"PM25"}).drop(columns=["Sample Duration"])
pm25_non_frm_fem=pm25_non_frm_fem[pm25_non_frm_fem["Sample Duration"]=="24-HR BLK AVG"]\
    .rename(columns={"Arithmetic Mean":"PM25_non"}).drop(columns=["Sample Duration"])
pmc=pmc[pmc["Sample Duration"]=="24 HOUR"]\
    .rename(columns={"Arithmetic Mean":"PMC"}).drop(columns=["Sample Duration"])
```

```python
pm10_g=pm10.groupby(by=["Latitude","Longitude","Date Local"]).mean().reset_index()
pm25_frm_fem_g=pm25_frm_fem.groupby(by=["Latitude","Longitude","Date Local"]).mean().reset_index()
pm25_non_frm_fem_g=pm25_non_frm_fem.groupby(by=["Latitude","Longitude","Date Local"]).mean().reset_index()
pmc_g=pmc.groupby(by=["Latitude","Longitude","Date Local"]).mean().reset_index()
```

```python
type1,type2,type3=[],[],[]
for i in range(len(summ1)):
    lat=summ1.index[i][0]
    lon=summ1.index[i][1]
    #print(len(stations[(stations["Latitude"]==lat)&(stations["Longitude"]==lon)]["Land Use"]))
    if len(stations[(stations["Latitude"]==lat)&(stations["Longitude"]==lon)]["Land Use"])==1:
        type1.append(stations[(stations["Latitude"]==lat)&(stations["Longitude"]==lon)]["Land Use"].values[0])
    else:
        type1.append(np.nan)
    if len(stations[(stations["Latitude"]==lat)&(stations["Longitude"]==lon)]["Location Setting"])==1:
        type2.append(stations[(stations["Latitude"]==lat)&(stations["Longitude"]==lon)]["Location Setting"].values[0])
    else:
        type2.append(np.nan)
    if len(stations[(stations["Latitude"]==lat)&(stations["Longitude"]==lon)]["Met Site Type"])==1:
        type3.append(stations[(stations["Latitude"]==lat)&(stations["Longitude"]==lon)]["Met Site Type"].values[0])
    else:
        type3.append(np.nan)
summ1["Land Use"],summ1["Location Setting"],summ1["Met Site Type"]=type1,type2,type3
summ1
```
